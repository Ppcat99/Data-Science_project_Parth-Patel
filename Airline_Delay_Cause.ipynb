```python
# This code will read the excel file above and get first five values
# First five value to see the which year it goes up to
df = pd.read_csv('Airline_Delay_Cause.csv')
df.head()
```

```python
# For tail to see which year it start from?
df.tail()
```

```python
# This code is used to combine 'year' and 'month' into one cell and name it
# 'date'
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))
df.head()
```

```python
# We removed the columns year and month because they were combined and put in
# on column and named date
df = df.drop(['year', 'month'], axis=1)
df.head()
```

```python
# Get the unique carrier names and sort them for easier reading
unique_carriers = sorted(df['carrier_name'].unique())

# Print the list of unique carriers
# This list of unique carriers is to find two different names in one airline
# company
print("\nList of Unique Airline Carriers:")
for carrier in unique_carriers:
    print(carrier)
```

```python
# Create a dictionary of replacements: key is the old name, value is the new name
# Used the list above to make the name one
replacements = {
    'American Airlines Inc.': 'American Airlines Network',
    'Delta Air Lines Inc.': 'Delta Air Lines Network',
    'Southwest Airlines Co.': 'Southwest Airlines',
    'Spirit Air Lines': 'Spirit Airlines',
    'United Air Lines Inc.': 'United Air Lines Network',
    'Hawaiian Airlines Inc.': 'Hawaiian Airlines',
    'Alaska Airlines Inc.': 'Alaska Airlines Network',
    'Frontier Airlines Inc.': 'Frontier Airlines',
    'ExpressJet Airlines Inc.': 'ExpressJet Airlines LLC',
    'Delta Air Lines Inc.': 'Delta Air Lines Network',
    'Commutair Aka Champlain Enterprises, Inc.': 'CommuteAir LLC dba CommuteAir',
}

# Apply the replacements to the 'carrier_name' column
df['carrier_name'] = df['carrier_name'].replace(replacements)

# Now, if you check the unique names again, these inconsistencies will be resolved.
print(sorted(df['carrier_name'].unique()))
```

```python
# This is a target value which is compared with the new 'date' column
# Use any column header with values in them not 'date'
target_variable = 'arr_delay'
```

```python
# Aggregate the data monthly to create a single time series
monthly_data = df.groupby(pd.Grouper(key='date', freq='ME'))[target_variable].sum().reset_index()
monthly_data.columns = ['date', 'value']
monthly_data.set_index('date', inplace=True)
```

```python
# 2. Feature Engineering for Time Series
# Create lagged features and a rolling mean
def create_features(df, lags=[1, 2, 3], rolling_windows=[3, 6]):
    df = df.copy()
    for lag in lags:
        df[f'lag_{lag}'] = df['value'].shift(lag)
    for window in rolling_windows:
        df[f'rolling_mean_{window}'] = df['value'].shift(1).rolling(window).mean()
    return df

monthly_data = create_features(monthly_data)
```

```python
# Drop rows with NaN values created by lagging and rolling mean
monthly_data.dropna(inplace=True)
```

```python
# 3. Splitting the data
# Use the most recent data for testing
train_size = int(len(monthly_data) * 0.8)
train = monthly_data.iloc[:train_size]
test = monthly_data.iloc[train_size:]
```

```python
# Continue Splitting the data to X and Y_train, X and Y_test
X_train = train.drop('value', axis=1)
y_train = train['value']
X_test = test.drop('value', axis=1)
y_test = test['value']
```

```python
# Importing packages of numpy, xgboost, and matplotlib
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
```

```python
# 4. XGBoost Model Training
model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1)
model.fit(X_train, y_train)
```

```python
# 5. Making Predictions
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)
```

```python
# 6. Forecasting future values
# To forecast future values, we need to create a future dataframe and fill it with lagged features
future_months = 13
last_date = monthly_data.index[-1]
future_dates = pd.date_range(start=last_date, periods=future_months + 1, freq='M')[1:]
```

```python
all_values = pd.concat([monthly_data['value'], pd.Series(dtype='float64')])
future_predictions = []
```

```python
# Iteratively predict future values
future_predictions = [] # Initialize future_predictions list
last_values = monthly_data['value'].copy() # Initialize last_values with existing data

for i in range(future_months):
    # Get the last few values to create new features for the next prediction
    if i == 0:
        # Use the actual last values from the dataset for the first prediction
        current_lags = [last_values.iloc[-lag] for lag in [1, 2, 3]]
        current_rolling_mean = last_values.iloc[-6:].mean()
    else:
        # For subsequent predictions, use the predicted values
        current_lags = [future_predictions[i-lag] if i-lag >= 0 else last_values.iloc[-1-(i-lag)] for lag in [1, 2, 3]]
        # Combine last_values and future_predictions for rolling mean calculation
        combined_values = pd.concat([last_values, pd.Series(future_predictions)])
        current_rolling_mean = combined_values.rolling(6).mean().iloc[-1]


    new_features = {
        'lag_1': current_lags[0],
        'lag_2': current_lags[1],
        'lag_3': current_lags[2],
        'rolling_mean_3': pd.concat([last_values, pd.Series(future_predictions)]).rolling(3).mean().iloc[-1],
        'rolling_mean_6': pd.concat([last_values, pd.Series(future_predictions)]).rolling(6).mean().iloc[-1]
    }

    new_features_df = pd.DataFrame([new_features], index=[future_dates[i]])


    # Predict the next value and append it to future_predictions
    next_prediction = model.predict(new_features_df)[0]
    future_predictions.append(next_prediction)

# Convert the list of future predictions to a numpy array
future_predictions = np.array(future_predictions)
```

```python
# Predict the next value
next_prediction = model.predict(new_features_df)[0]
```

```python
# Create a final DataFrame for the forecast
forecast_df = pd.DataFrame({'date': future_dates, 'value': future_predictions})
display(forecast_df.head())
```

```python
# 7. Visualization
# Filter data for visualization from 2000 onwards
start_date = '2000-01-01'
monthly_data_filtered = monthly_data[monthly_data.index >= start_date]
train_filtered = train[train.index >= start_date]
test_filtered = test[test.index >= start_date]
y_test_filtered = y_test[y_test.index >= start_date]
train_predictions_filtered = train_predictions[train.index >= start_date]
test_predictions_filtered = test_predictions[test.index >= start_date]


plt.figure(figsize=(15, 7))
plt.plot(monthly_data_filtered.index, monthly_data_filtered['value'], label='Actual Data')
plt.plot(train_filtered.index, train_predictions_filtered, label='Training Predictions', linestyle='--')
plt.plot(test_filtered.index, y_test_filtered, label='Actual Test Data')
plt.plot(test_filtered.index, test_predictions_filtered, label='Test Predictions', linestyle='--')
plt.plot(forecast_df.index, forecast_df['value'], label='Future Forecast', linestyle=':', color='red')
plt.title(f'XGBoost Time Series Forecasting for {target_variable}')
plt.xlabel('Date')
plt.ylabel(f'Total {target_variable}')
plt.legend()
plt.grid(True)
plt.show()

print("\nFuture Forecast (Next 12 months):")
print(forecast_df)
```
