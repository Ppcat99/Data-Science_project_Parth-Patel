```python
# Libraries to help with data visulaization like creating graphs and charts
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Clean Columns names for easier use to change to covert to lowercase
df.columns = df.columns.str.replace(' ', '_').str.lower()
print("Cleaned Columns:", df.columns.tolist())
```

```python
# Question 1: How are the "math score" distributed among students?
# To answer this question we are making a code of a histogram
plt.figure(figsize=(8,5))
sns.histplot(df['math_score'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()

# Distrubution is left skewed, so most of student are making between 60 to 70%
# in their math test score.
```

```python
# Question 2: What is the breakdown of "parental level of education"
# The purpose of this question is to see the education background of the
# student's parents
plt.figure(figsize=(10,6))

education_order = ['some high school', 'high school', 'some college',
                   "associate's degree", "bachelor's degree", "master degree"]

sns.countplot(data=df, y='parental_level_of_education', order=education_order,
              palette='magma')
plt.title('Distribution of Parental level of Education')
plt.xlabel('Number of students')
plt.ylabel('Parental Education Level')
plt.show()
# According to the bar chart below most of the students's parents have some
# college and associate degree
```

```python
# Question 3: How does gender infludence student performance across 'math score',
# 'reading score', and 'writing score'

# 3. Generate Violin Plots for Gender vs. Scores

# Set a style for the plots for better aesthetics
sns.set_style("whitegrid")

# Create a figure with 3 subplots (1 row, 3 columns) for comparison
plt.figure(figsize=(20, 7)) # Adjust figure size for better readability

# Plot 1: Math Score Distribution by Gender using Violin Plot
plt.subplot(1, 3, 1) # (rows, columns, plot_number)
sns.violinplot(data=df, x='gender', y='math_score', palette={'male': '#98c1d9', 'female': '#ee6c4d'}, inner='quartile')
plt.title('Math Score Distribution by Gender (Violin Plot)', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Math Score', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(0, 100) # Consistent y-axis

# Plot 2: Reading Score Distribution by Gender using Violin Plot
plt.subplot(1, 3, 2)
sns.violinplot(data=df, x='gender', y='reading_score', palette={'male': '#98c1d9', 'female': '#ee6c4d'}, inner='quartile')
plt.title('Reading Score Distribution by Gender (Violin Plot)', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Reading Score', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(0, 100) # Consistent y-axis

# Plot 3: Writing Score Distribution by Gender using Violin Plot
plt.subplot(1, 3, 3)
sns.violinplot(data=df, x='gender', y='writing_score', palette={'male': '#98c1d9', 'female': '#ee6c4d'}, inner='quartile')
plt.title('Writing Score Distribution by Gender (Violin Plot)', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Writing Score', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(0, 100) # Consistent y-axis

plt.tight_layout() # Adjusts subplot parameters for a tight layout
plt.show()

# Males generally show a slight advantage in Math
# Female demonstrate a noticeable advantage in reading
# Female show a significent and string advantage in writing
```

```python
# Question 4: Does the type of lunch (Standard or free/reduced) impact student
# performance across the subjects?

# Reading score vs Lunch
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='lunch', y='reading_score', palette='pastel')
plt.xlabel('Lunch Type', fontsize=12)
plt.ylabel('Reading Score', fontsize=12)
plt.title('Reading Score distribution by Lunch Time', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
# Math Score vs lunch
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='lunch', y='math_score', palette='pastel')
plt.xlabel('Lunch Type', fontsize=12)
plt.ylabel('Math Score', fontsize=12)
plt.title('Math Score distribution by Lunch Time', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
# Writing Score vs lunch
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='lunch', y='writing_score', palette='pastel')
plt.xlabel('Lunch Type', fontsize=12)
plt.ylabel('Writing Score', fontsize=12)
plt.title('Writing Score distribution by Lunch Time', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
# Conclusion of three boxplot is that if the student are given standard lunch
# they are highly capable to do well during test.
```

```python
df['average_score'] = (df['math_score'] + df['reading_score'] + df['writing_score']) / 3
df.head()
```

```python
# Question 5: Is there a relationship between 'lunch' type and 'test preparation
# course completion?

plt.figure(figsize=(9,6))
sns.countplot(data=df, x='lunch', hue='test_preparation_course', palette='viridis')
plt.title('Test Preparation Course Completion by lunch type', fontsize=16)
plt.xlabel('Lunch Type', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Test Prep Course')
plt.tight_layout()
plt.show()

# Wow, according to this plot we can see that people who did not took the prep
# course were enjoying nice lunch vs people who were completing their prep
# course during lunch time.
```
